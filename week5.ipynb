```python
target_ip = '8.8.8.8'

try:
    host_info = api.host(target_ip)
    print(f'organization:{host_info["org"]}')

    for item in host_info.get('data',[]):
        port = item.get('port')
        banner=item.get('data','N/A').split('\n')[0]
        print(f"port{port}|banner{banner}")
except shodan.APIError as e:
    print(e)
```

```python
query = 'product:OpenSSH'
FACETS = [('org', 10)]

try:
    summary = api.count(query, facets=FACETS)
    print(f'global OpenSSH usage total: {summary["total"]}')

    organization = summary.get('facets', {}).get('org', [])
    print('top 10 organizations using OpenSSH:')
    for item in organization:
        print(f"{item['value']}: {item['count']} hosts")
except shodan.APIError as e:
    print(e)
```
