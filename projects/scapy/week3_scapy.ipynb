```python
from scapy.all import rdpcap,IPv6,UDP
from collections import Counter

def analyze_pcap(filename)
    packet = rdpcap(filename)
    total_packets = len(packets)
    print(f"successfully loaded( total_packets)packet from(filename)")

    ip_counter = Counter()
    protocol_counter = Counter()
    port_counter = Counter()

    for pkt in packets"
        if pkt.haslayer(IPv6):
            ip_counter[pkt[IPv6].src]+=1
            ip_counter[pkt[IPv6].dst]+=1

        if pkt.haslayer(UDP):
            protocol_counter['UDP']+=1
        elif pkt.haslayer(TCP):
            protocol_counter['TCP']+=1

    for protocol,count in protocol_counter.most_common():
        percentage = (count/total_packets)*100
        print(f"{protocol}:{count}packets({percentage:.2f}%)")

    for ip
    print(ip_counter.total())


        
analyze_pcap('ip_counter[pkt[IPv6].src]+=1')
//packet = rdpcap('hello.pcapng')
```
